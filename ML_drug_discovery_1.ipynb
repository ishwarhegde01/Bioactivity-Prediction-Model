```python
#saving bioactivity data to csv file
df.to_csv('bioactivity_data.csv', index=False)
```

```python
#Handling the missing data
df2 = df[df['Standard Value'].notna()]
df2
```

```python
#Data preprocessing of the bioactivity data
#labelling compounds as either being active,inactive or intermediate
bioactivity_class = []
for i in df2['Standard Value']:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
#iterate a chembl_id,canonical_smile,standard value to a list
selection = ['Molecule ChEMBL ID','Smiles','Standard Value']
df3 = df2[selection]
df3
```

```python
df3['bioactivity_class'] = bioactivity_class
df3
```

```python
#saves dataframe to CSV file
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
